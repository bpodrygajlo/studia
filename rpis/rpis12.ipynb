załóżmy, że średnie w grupach (grupowanie po dniu) są równe

```python
for name,x in df.groupby('dzien'):
    print "Dzień", name, ", średnia:", x['razem'].mean()
```

Średnie nie są równe. Przeprowadzimy ANOVA.

```python
srednia_calkowita = df['razem'].mean()
srednia_calkowita
```

Suma kwadratów róznic X_i - srednia_calkowita

```python
SST = ((df['razem'] - srednia_calkowita)**2).sum()
SST
```

```python
SSTR = 0
for dzien,x in df.groupby('dzien'):
    SSTR += len(x['razem'])*(x['razem'].mean() - srednia_calkowita)**2
SSTR
```

```python
SSE = 0
for dzien,x in df.groupby('dzien'):
    SSE += ((x['razem']-x['razem'].mean())**2).sum()
SSE
```

SST = SSTR + SSE

```python
print SST
print SSTR+SSE
```

```python
N = len(df['razem'])
c = len(df.groupby('dzien'))
print "N,c:", N,c
```

```python
MST = SST/(N-1)
print "MST:", MST
MSTR = SSTR/(c-1)
print "MSTR:", MSTR
MSE = SSE/(N-c)
print "MSE:", MSE
F = MSTR/MSE
print "F:", F
from scipy.stats import f
df1 = c-1
df2 = N-c
f.cdf(F,df1,df2)
```

##Zadnie 2

```python
def anova1(df,groups,plain):
    print "Test 1. hipotezy"

    
    for name,x in groups:
        print name, ", średnia:", x.mean()
    
    
    srednia_calkowita = plain.mean()
    print "Calkowita srednia:", srednia_calkowita
    
    SST = ((plain - srednia_calkowita)**2).sum()
    print "SST:", SST
    
    SSTR = 0
    for dzien,x in groups:
        SSTR += len(x)*(x.mean() - srednia_calkowita)**2
    print "SSTR", SSTR
    
    SSE = 0
    for dzien,x in groups:
        SSE += ((x-x.mean())**2).sum()
    print "SSE:", SSE
    
    print "SST = SSTR + SSE"
    print "SST", SST
    print "SSTR+SSE", SSTR+SSE
    
    N = len(plain)
    c = len(groups)
    print "N,c:", N,c
    
    MST = SST/(N-1)
    print "MST:", MST
    MSTR = SSTR/(c-1)
    print "MSTR:", MSTR
    MSE = SSE/(N-c)
    print "MSE:", MSE
    F = MSTR/MSE
    print "F:", F
    from scipy.stats import f
    df1 = c-1
    df2 = N-c
    p = f.cdf(F,df1,df2)
    if p > 0.05:
        print "Hipoteza odrzucona z 95% pewnością"
    else:
        print "Hipoteza nie została odrzucona"
```

```python
groups = df.groupby('poziom')['razem']
plain = df['razem']
anova1(df,groups,plain)
```
