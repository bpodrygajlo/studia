## Zadanie 2.

$$\Theta = (X X^T)^{-1} X Y^T$$

```python
X = df.drop(['punkty','zawodnik','kraj','miejsce'], axis = 1).values
X = X.T
X
```

```python
Y = df['punkty'].values
Y
```

```python
Theta = np.linalg.inv(X.dot(X.T)).dot(X).dot(Y.T)
Theta
```

Kolumna predict - wartości wyliczone za pomocą modelu  
Kolumna values - punkty

```python
res = pd.DataFrame()
res['predict'] = Theta.T.dot(X)
res['values'] = Y
res
```
